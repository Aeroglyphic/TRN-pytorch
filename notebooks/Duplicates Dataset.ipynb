```python
import pickle

duplicates = pickle.load(open('duplicates_mit.pkl', 'rb'))
```

```python
MAX_LEVEL = 10
level = 0

def add_neighbours(item, keys):
    global level
    level += 1
    if level > MAX_LEVEL:
        return
    for x in keys:
        x = str(x)
        n_neighbours = duplicates.get(x, [])
        n_neighbours = [str(x) for x in n_neighbours]
        item.add(x)
        item.update(n_neighbours)
        add_neighbours(item, n_neighbours)

results = []
for i, (gif_id, neighbours) in enumerate(duplicates.items()):
    level = 0
    gif_id = str(gif_id)
    neighbours = [str(x) for x in neighbours]
    if i % 1000 == 0:
        print(i)
    new_item = set(neighbours)
    new_item.add(gif_id)
    add_neighbours(new_item, neighbours)
    for i, x in enumerate(results):
        if len(x & new_item) > 0:
            results[i].update(new_item)
            new_item = set()
            break
    if len(new_item) > 0:
        results.append(new_item)
```

```python
import pickle

pickle.dump(results, open('dupl.pkl', 'wb'))
```
